```python
from diffusers import DiffusionPipeline, StableDiffusionXLImg2ImgPipeline
import torch

model = "stabilityai/stable-diffusion-xl-base-1.0"
pipe = DiffusionPipeline.from_pretrained(
    model,
    torch_dtype=torch.float16,
)
pipe.to("cuda")
pipe.load_lora_weights("/content/Abishek_ND_3/", weight_name="pytorch_lora_weights.safetensors")

refiner = StableDiffusionXLImg2ImgPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-refiner-1.0",
    torch_dtype=torch.float16,
)
refiner.to("cuda")
```

```python
prompt = "a close up portrait of sks woman, realistic, gorgeous, 4k, 8k, uhd, cinematic, masterpiece, dramatic, black dress, beautiful lighting, sharp details."

for seed in range(10):
    generator = torch.Generator("cuda").manual_seed(seed)
    image = pipe(prompt=prompt, generator=generator, num_inference_steps=25)
    image = image.images[0]
    image.save(f"/content/drive/MyDrive/DL_Project/Abishek_ND_3_images1/{seed}.png")
    image = refiner(prompt=prompt, generator=generator, image=image)
    image = image.images[0]
    image.save(f"/content/drive/MyDrive/DL_Project/Abishek_ND_3_images_refined/{seed}.png")
```
